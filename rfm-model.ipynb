## Data Preparation

**The RFM model analyses the data on three factors:**
1. Recency: How recent the customer visited the website?
2. Frequency: How frequent the customer visited the website?
3. Monetary: How much money did the customer spend at our website?

```python
# Monetary

retail['Amount'] = retail['Quantity'] * retail['UnitPrice']
rfm_m = retail.groupby('CustomerID')['Amount'].sum()
rfm_m = rfm_m.reset_index()
rfm_m.head()
```

```python
# Frequency

rfm_f = retail.groupby('CustomerID')['InvoiceNo'].count()
rfm_f = rfm_f.reset_index()
rfm_f.columns = ['CustomerID', 'Frequency']
rfm_f.head()
```

```python
rfm = pd.merge(rfm_m, rfm_f, on='CustomerID', how='inner')
rfm.head()
```

```python
# Recency

# Convert to datetime to proper datetime

retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'], format='%d-%m-%Y %H:%M')
```

```python
# Compute the maximum date so that you can know the last transaction date

max_date = max(retail['InvoiceDate'])
max_date
```

```python
# Compute the difference between max date and transaction date

retail['Diff'] = max_date - retail['InvoiceDate']
retail.head()
```

```python
# Compute the last transaction date to get the recency of customers

rfm_p = retail.groupby('CustomerID')['Diff'].min()
rfm_p = rfm_p.reset_index()
rfm_p.head()
```

```python
# Extract the number of days only

rfm_p['Diff'] = rfm_p['Diff'].dt.days
rfm_p.head()
```

```python
# Merge the dataframe to get the final RFM dataframe

rfm = pd.merge(rfm, rfm_p, on='CustomerID', how='inner')
rfm.columns = ['CustomerID', 'Amount', 'Frequency', 'Recency']
rfm.head()
```

**There are two types of outliers and we will treat it as it can skew our dataset.**

- Statistical
- Domain Specific

```python
# Outlier analysis of Amount, Frequency, and Recency

attributes = ["Amount", "Frequency", "Recency"]
plt.rcParams['figure.figsize'] = [10, 8]
sns.boxplot(data = rfm[attributes], orient="v", palette="Set2", whis=1.5, saturation=1,
            width = 0.7)
plt.title("Outliers Variable Distribution", fontsize = 14, fontweight='bold')
plt.ylabel("Range", fontweight = "bold")
plt.xlabel("Attributes", fontweight = "bold")
```

```python
# Removing statistical outliers for Amount
Q1 = rfm.Amount.quantile(0.05)
Q3 = rfm.Amount.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Amount >= Q1 - 1.5 * IQR) & (rfm.Amount <= Q3 + 1.5*IQR)]

# Removing statistical outliers for Recency
Q1 = rfm.Recency.quantile(0.05)
Q3 = rfm.Recency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Recency >= Q1 - 1.5*IQR) & (rfm.Recency <= Q3 + 1.5*IQR)]

# Removing statistical outliers for Frequency
Q1 = rfm.Frequency.quantile(0.05)
Q3 = rfm.Frequency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Frequency >= Q1 - 1.5*IQR) & (rfm.Frequency <= Q3 + 1.5*IQR)]
```

**Rescaling Attributes**
It is extremely important to rescale attributes so that they are of comparable scale. There are two common ways for rescaling:
1. Min-Max Scaling
2. Standardization (mean-0, sigma-1)

Here we are going to use standard scaling

```python
# Rescaling the attributes

rfm_df = rfm[['Amount', 'Frequency', 'Recency']]

# Instantiate
scaler = StandardScaler()

# fit_transform
rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

```python
rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['Amount', 'Frequency', 'Recency']
rfm_df_scaled.head()
```

## Building the Model

K-Means Clustering

K-Means clustering works as follows:
- First we initialize k points called means randomly
- We categorize each item to its closest mean and we update the mean's coordinates, which are the averages of the items categorized.
- We repeat the process for a given number of iterations and at the end we have our clusters.

```python
kmeans = KMeans(n_clusters = 4, max_iter = 50)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

### Finding the optimal number of clusters
We can use the elbow curve to find the optimal number of clusters.

```python
# Elbow-curve/SSD

ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters = num_clusters, max_iter = 50)
    kmeans.fit(rfm_df_scaled)
    
    ssd.append(kmeans.inertia_)

plt.plot(ssd);
```

### Silhouette Score
silhouette score = $\large\frac{p\space -\space q}{max(p,q)} $

p is the mean distance to the closest points that is not part of the cluster
q is the intra mean distance of the points in its own cluster

- The value of the silhouette score ranges from -1 to 1.
- A score closer to 1 indicates that the data point is very similar to other data points in the cluster
- A score closer to -1 indicates that the data point is not very similar to other data points in the cluster

```python
# Silhouette Analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9]

for num_cluster in range_n_clusters:
    
    #initialize kmeans
    kmeans = KMeans(n_clusters = num_cluster, max_iter = 50)
    kmeans.fit(rfm_df_scaled)
    
    cluster_labels = kmeans.labels_
    
    #silhouette score
    silhouette_avg = silhouette_score(rfm_df_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_cluster, silhouette_avg))
```

```python
# Final model with a k of 3
kmeans = KMeans(n_clusters = 3, max_iter = 50)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

```python
rfm['ClusterID'] = kmeans.labels_
rfm.head()
```

```python
# Box plot to visualize Cluster Id vs Amount

sns.boxplot(x = 'ClusterID', y = 'Amount', data=rfm);
```

```python
# Box plot to visualize Cluster Id vs Frequency

sns.boxplot(x = 'ClusterID', y = 'Frequency', data = rfm);
```

```python
# Box plot for Cluster Id vs Recency

sns.boxplot(x = 'ClusterID', y = 'Recency', data = rfm);
```

## Hierarchical Clustering

Hierarchical clustering involves creating clusters that have a predetermined ordering from top to bottom. For example all the files and folders on the hard disk is organized as a hierarchy. There are two types of clustering:
- Agglomerative
- Divisive

### Single Linkage
- In single linkage hierarchical clustering, the distance between the two clusters is defined as the shortest distance between two points in a cluster.


```python
# Single Linkage

mergings = linkage(rfm_df_scaled, method = 'single', metric='euclidean')
dendrogram(mergings)
plt.show();
```

### Complete Linkage
- In complete linkage hierarchical clustering, the distance between two clusters is defined by the longest distance between the points in the two clusters.

```python
mergings = linkage(rfm_df_scaled, method = 'complete', metric='euclidean')
dendrogram(mergings)
plt.show()
```

### Average Linkage
- In average linkage hierarchical clustering, the distance between two clusters is defined as the average distance between each point in one cluster to every point in the other clusters.

```python
# Average Linkage

mergings = linkage(rfm_df_scaled, method='average', metric='euclidean')
dendrogram(mergings)
plt.show()
```

**Cutting the Dendrogram based on K**

```python
# 3 cluster lables

cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
# Assign cluster labels

rfm['ClusterLabels'] = cluster_labels
rfm.head()
```

```python
# Plot Id vs Amount

sns.boxplot(x = 'ClusterLabels', y = 'Amount', data = rfm);
```

```python
# Plot Cluster Id vs Frequency

sns.boxplot(x = 'ClusterLabels', y = 'Frequency', data=rfm);
```

```python
# Plot Id vs Recency

sns.boxplot(x = 'ClusterLabels', y = 'Recency', data = rfm);
```
